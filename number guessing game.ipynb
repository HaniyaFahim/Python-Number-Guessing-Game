```python
import threading
import tkinter as tk
from tkinter import *


win = tk.Tk()

win.title('Guess The Number 10')
win.geometry('400x400')
win.configure(bg="#333")


mylabel = tk.Label(win, text="Guess the number 10", bg="pink" ,font=(  'DilleniaUPC',26,'bold'))
mylabel.grid(row=0,column=1)

#button 1 functions

def click_me1():
    B1.configure(text="Looser" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="30"  ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_mei).start()
    
def click_mei():
    B1.configure(text="" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Now Let's Play Again" ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_me0).start()
    
def click_me0():
    B1.configure(text="Guess 10" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Guess the number 10"  ,font=(  'DilleniaUPC',26,'bold'))
    
#button 2 functions

def click_me2():
    B2.configure(text="Looser" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="40" ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_me).start()
    
def click_me():
    B2.configure(text="" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Now Let's Play Again" ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_mel).start()
    
def click_mel():
    B2.configure(text="Guess 10" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Guess the number 10"  ,font=(  'DilleniaUPC',26,'bold'))
    
#button 3 functions
    
def click_me3():
    B3.configure(text="Loser" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="20"  ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_met).start()
    
def click_met():
    B3.configure(text="" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Now Let's Play Again"  ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_meo).start()
    
def click_meo():
    B3.configure(text="Guess 10" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Guess the number 10"  ,font=(  'DilleniaUPC',26,'bold'))
    


#button 4 functions

def click_me4():
    B4.configure(text="Winner" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="10" ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_men).start()
    
def click_men():
    B4.configure(text="" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Now Let's Play Again"  ,font=(  'DilleniaUPC',26,'bold'))
    threading.Timer(2, click_mec).start()
    
def click_mec():
    B4.configure(text="Guess 10" ,font=(  'Gisha',16,'bold'))
    mylabel.configure(text="Guess the number 10"  ,font=(  'DilleniaUPC',26,'bold'))
    
    
    


B1 = Button(win, text="Guess 10", bg="yellow", command=click_me1  ,font=(  'Gisha',16,'bold'))
B1.grid(row=1,column=1)

B2 = Button(win, text="Guess 10", bg="yellow", command=click_me2 ,font=(  'Gisha',16,'bold'))
B2.grid(row=2,column=1)

B3 = Button(win, text="Guess 10", bg="yellow", command=click_me3 ,font=(  'Gisha',16,'bold'))
B3.grid(row=3,column=1)

B4 = Button(win, text="Guess 10", bg="yellow", command=click_me4 ,font=(  'Gisha',16,'bold'))
B4.grid(row=4,column=1)

win.mainloop()
```
